### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu'].value_counts()
```

```python
df['Cpu_company'] = df['Cpu'].apply(lambda row: row.split()[0])
df.sample(5)
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df.Memory.value_counts()
```

```python
# 512GB SSD +  1.0TB Hybrid  
def get_gb(row: str) -> str:
    gb = 0
    row_list = row.split()
    for el in row_list:
        if 'GB' in el:
            gb += int(el[:-2])
        elif 'TB' in el:
            gb += float(el[:-2]) * 1024
    return int(gb)

df['Memory_Amount'] = df['Memory'].apply(get_gb)
df.sample(5)
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
# 64GB Flash Storage +  1TB HDD

def get_type(row: str) -> str:
    type_memory = ''
    row_list = row.split()
    for el in row_list:
        if 'GB' not in el and 'TB' not in el and el not in type_memory:
            type_memory += el + ''
    return type_memory.rstrip().rstrip(' +')

df['Memory_Type'] = df['Memory'].apply(get_type)
df.sample(5)
```
