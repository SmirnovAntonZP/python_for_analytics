## Задание 1. Изменение и сохранение данных





### 1. Считать данные из файла kc_house_data.csv с помощью pandas.

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('./kc_house_data.csv')
df.sample()
```

### 2. Добавить новый признак price_per_sqft_living, который будет содержать среднюю стоимость за квадратный метр жилой площади (price / sqft_living).

```python
df['price_per_sqft_living'] = df['price']/df['sqft_living']
df.sample()
```

### 3. Создать новый признак age_of_house, который будет содержать возраст дома (текущий год минус год постройки yr_built).

```python
current_year = pd.to_datetime('today').year
df['age_of_house'] = current_year - df['yr_built']
df.sample()
```

### 4. Удалить признаки zipcode, lat, long и сохранить измененные данные в новый CSV файл modified_data.csv.

```python
new_df = df.drop(columns=['zipcode','lat','long'])
new_df.sample()
```

```python
new_df.to_csv('modified_data.csv',index=False)
```

## Задача 3. Анализ жилых и общих площадей




### 1. Считать данные из файла kc_house_data.csv.

```python
pd.set_option('display.max_columns', None)
df3 = pd.read_csv('./kc_house_data.csv')
df3.sample()
```

### 2. Построить DataFrame с колонками sqft_living, condition и price.

```python
df3[['sqft_living', 'condition', 'price']]
```

### 3. Определить среднюю стоимость за квадратный метр жилой площади для разных состояний домов и сохраните в текстовый файл.

```python
df3['Price_per_sqft'] = df3['price'] / df3['sqft_living']
df3.sample()
```

```python
mean_price_sqft_living_condition = df3.groupby('condition').agg({'Price_per_sqft': 'mean'})
mean_price_sqft_living_condition.head()
```

```python
with open('living_vs_condition_analysis.txt', 'w') as f:
    f.write(f'Средняя стоимость за квадратный метр жилой площади для разных состояний домов:{mean_price_sqft_living_condition}\n')
```

## Задача 4. Информация о клиентах и их покупках

### 1. Создайте датафрейм с покупками и сохраните его в переменную purchases.
purchases = pd.DataFrame({

```python
purchases = pd.DataFrame({ 'purchase_id': [2001, 2002, 2003, 2004, 2005], 
                          'client_id': [1001, 1002, 1003, 1001, 1005], 
                          'house_id': [1234567890, 9876543210, 4567891230, 5566778899, 1122334455] })
purchases.head()
```

### 2. Создайте датафрейм с информацией о клиентах и сохраните его в переменную clients_info.
clients_info = pd.DataFrame({
'client_id': [1001, 1002, 1003, 1004, 1005],
'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
'age': [30, 40, 35, 45, 50]
})

```python
clients_info = pd.DataFrame({ 'client_id': [1001, 1002, 1003, 1004, 1005], 
                             'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'], 
                             'age': [30, 40, 35, 45, 50] })
clients_info.head()
```

### 3. Присоедините информацию о клиентах к таблице purchases через метод join
и сохраните в переменную purchases_with_info.

```python
purchases_with_info = purchases.join(clients_info.set_index('client_id'), on = 'client_id')
purchases_with_info.head()
```

### 4. Присоедините информацию о клиентах к таблице purchases через метод
merge и сохраните в переменную purchases_with_info_merged.

```python
purchases_with_info_merged = purchases.merge(clients_info,left_on = 'client_id', right_on = 'client_id', how = 'left')
purchases_with_info_merged.head()
```
